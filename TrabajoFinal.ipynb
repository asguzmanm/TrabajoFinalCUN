**Relación entre repitencia y deserción por nivel educativo**

**Departamento**

```python
import pandas as pd

niveles = {
    "Primaria": "repitencia_primaria",
    "Secundaria": "repitencia_secundaria",
    "Media": "repitencia_media"
}

df_niveles = []

for nivel, col_rep in niveles.items():
    temp = df_depto[["año", "deserción", col_rep]].copy()
    temp.columns = ["año", "deserción", "repitencia"]
    temp["nivel"] = nivel
    df_niveles.append(temp)

df_niveles = pd.concat(df_niveles, ignore_index=True)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))

sns.scatterplot(
    data=df_niveles,
    x="repitencia",
    y="deserción",
    hue="nivel",
    s=80
)

sns.regplot(
    data=df_niveles,
    x="repitencia",
    y="deserción",
    scatter=False,
    color="black",
    line_kws={"linewidth":2}
)

plt.title("Relación entre repitencia y deserción escolar\npor nivel educativo")
plt.xlabel("Repitencia (%)")
plt.ylabel("Deserción (%)")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
from scipy.stats import spearmanr

resultados_corr = []

for nivel in df_niveles["nivel"].unique():
    sub = df_niveles[df_niveles["nivel"] == nivel]
    rho, pval = spearmanr(sub["repitencia"], sub["deserción"])
    resultados_corr.append([nivel, rho, pval])

corr_df = pd.DataFrame(
    resultados_corr,
    columns=["Nivel educativo", "Spearman ρ", "p-valor"]
)

corr_df
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import spearmanr

heat_data = []

for nivel in df_niveles["nivel"].unique():
    sub = df_niveles[df_niveles["nivel"] == nivel]
    rho, _ = spearmanr(sub["repitencia"], sub["deserción"])
    heat_data.append([nivel, rho])

heat_df = pd.DataFrame(heat_data, columns=["Nivel", "Spearman_rho"])
heat_df = heat_df.set_index("Nivel")

plt.figure(figsize=(4,3))
sns.heatmap(
    heat_df,
    annot=True,
    cmap="Blues",
    fmt=".2f",
    vmin=0, vmax=1
)
plt.title("Correlación repitencia–deserción\npor nivel educativo (Spearman)")
plt.tight_layout()
plt.show()
```

**Análisis de correlación estratificado por IPM**

**Departamental**

```python
df_niveles_ipm = df_niveles.merge(
    df_depto[['año', 'ipm']],
    on='año',
    how='left'
)

df_niveles_ipm['grupo_ipm'] = pd.qcut(
    df_niveles_ipm['ipm'],
    q=3,
    labels=['IPM bajo', 'IPM medio', 'IPM alto']
)
```

```python
from scipy.stats import spearmanr

resultados = []

for nivel in df_niveles_ipm['nivel'].unique():
    for grupo in df_niveles_ipm['grupo_ipm'].unique():
        sub = df_niveles_ipm[
            (df_niveles_ipm['nivel'] == nivel) &
            (df_niveles_ipm['grupo_ipm'] == grupo)
        ]

        rho, p = spearmanr(
            sub['repitencia'],
            sub['deserción'],
            nan_policy='omit'
        )

        resultados.append([nivel, grupo, rho, p])

corr_ipm_df = pd.DataFrame(
    resultados,
    columns=['Nivel educativo', 'Grupo IPM', 'Spearman ρ', 'p-valor']
)

corr_ipm_df
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

pivot = corr_ipm_df.pivot(
    index='Nivel educativo',
    columns='Grupo IPM',
    values='Spearman ρ'
)

plt.figure(figsize=(6,4))
sns.heatmap(
    pivot,
    annot=True,
    cmap='Blues',
    fmt='.2f'
)
plt.title('Correlación repitencia–deserción\nsegún nivel educativo y contexto territorial (IPM)')
plt.tight_layout()
plt.show()
```

# **Desarrollo del Modelo**

Riesgo alto de deserción = 1
Riesgo bajo = 0

Se usa el percentil 75 de la deserción.

```python
import numpy as np

umbral = df_muni['deserción'].quantile(0.75)

df_muni['riesgo_desercion'] = (
    df_muni['deserción'] >= umbral
).astype(int)

df_muni['riesgo_desercion'].value_counts(normalize=True)
```

Selección de variables predictoras

```python
features = [
    'repitencia_primaria',
    'repitencia_secundaria',
    'repitencia_media',
    'ipm'
]

X = df_muni[features]
y = df_muni['riesgo_desercion']
```

Train / Test split

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.3,
    random_state=42,
    stratify=y
)
```

**Regresión Logística**

```python
from sklearn.linear_model import LogisticRegression

logit = LogisticRegression(max_iter=1000)
logit.fit(X_train, y_train)

y_pred_logit = logit.predict(X_test)
y_prob_logit = logit.predict_proba(X_test)[:,1]
```

```python
from sklearn.metrics import (
    accuracy_score, precision_score, recall_score,
    f1_score, confusion_matrix, roc_auc_score
)

print("Accuracy:", accuracy_score(y_test, y_pred_logit))
print("Precision:", precision_score(y_test, y_pred_logit))
print("Recall:", recall_score(y_test, y_pred_logit))
print("F1:", f1_score(y_test, y_pred_logit))
print("ROC-AUC:", roc_auc_score(y_test, y_prob_logit))

confusion_matrix(y_test, y_pred_logit)
```

**Random Forest**

```python
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier(
    n_estimators=300,
    random_state=42,
    class_weight='balanced'
)

rf.fit(X_train, y_train)

y_pred_rf = rf.predict(X_test)
y_prob_rf = rf.predict_proba(X_test)[:,1]
```

```python
print("Accuracy:", accuracy_score(y_test, y_pred_rf))
print("Precision:", precision_score(y_test, y_pred_rf))
print("Recall:", recall_score(y_test, y_pred_rf))
print("F1:", f1_score(y_test, y_pred_rf))
print("ROC-AUC:", roc_auc_score(y_test, y_prob_rf))

confusion_matrix(y_test, y_pred_rf)
```

**Comparación**
